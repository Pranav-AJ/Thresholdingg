```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('leaf.jpeg')
plt.imshow(image)
plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.show()
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the original grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Global thresholding
ret_global, th_global = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Display the global thresholding result
plt.imshow(th_global, cmap='gray')
plt.title('Global Thresholding (v=127)')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Adaptive thresholding (Gaussian)
th_adaptive = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                    cv2.THRESH_BINARY, 11, 2)

# Display the adaptive thresholding result
plt.imshow(th_adaptive, cmap='gray')
plt.title('Adaptive Gaussian Thresholding')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Otsu's thresholding
ret_otsu, th_otsu = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the Otsu thresholding result
plt.imshow(th_otsu, cmap='gray')
plt.title("Otsu's Thresholding")
plt.xticks([]), plt.yticks([])
plt.show()
```

Note :
1. Global Thresholding:
In global thresholding, a single threshold value is applied to the entire image. Any pixel value above the threshold (127 in our case) is set to 255 (white), and any value below it is set to 0 (black).

Effect:

The result is a binary image (black and white).
Areas of the image that are brighter than the threshold will turn completely white, while darker areas will turn black.

After Global Thresholding:

All pixel values below 127 are converted to 0 (black).
All pixel values above 127 are converted to 255 (white).
This works well in images with uniform lighting but struggles with images where the contrast is not consistent.

```python
2. Adaptive Thresholding:
In adaptive thresholding, different threshold values are calculated for small regions of the image (based on the pixel values in their neighborhood). This means the threshold can vary across different parts of the image, allowing it to handle varying lighting conditions better.

Effect:

The threshold adapts to local conditions, making it effective for images with non-uniform illumination.
It divides the image into smaller regions and applies a different threshold to each region based on the local mean or Gaussian weighted sum.
```
